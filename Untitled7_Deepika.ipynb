```python
import pandas as pd

df = pd.read_csv('Indian Airlines.csv')
display(df.head())
print(df.shape)
```

```python
# Handle missing values
for col in df.columns:
    if df[col].isnull().any():
        if pd.api.types.is_numeric_dtype(df[col]):
            df[col].fillna(df[col].mean(), inplace=True)
        else:
            df[col].fillna(df[col].mode()[0], inplace=True)

# Remove duplicates
df.drop_duplicates(inplace=True)

# Outlier detection and treatment
for col in ['duration', 'days_left', 'price']:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]

# Categorical value standardization
def standardize_city(city):
    return city.title()

for col in ['source_city', 'destination_city']:
    df[col] = df[col].apply(standardize_city)
display(df.head())
```

```python
# Examine data types
print(df.dtypes)

# Determine DataFrame shape
print("\nDataFrame Shape:", df.shape)

# Descriptive statistics for numerical columns
numerical_cols = df.select_dtypes(include=['number'])
print("\nDescriptive Statistics for Numerical Columns:")
display(numerical_cols.describe())

# Unique values and frequencies for categorical columns
categorical_cols = df.select_dtypes(exclude=['number'])
for col in categorical_cols.columns:
    print(f"\nUnique values and frequencies for '{col}':")
    display(categorical_cols[col].value_counts())
```

```python
# Calculate summary statistics for numerical features
numerical_features = ['duration', 'days_left', 'price']
summary_stats = df[numerical_features].describe()
print("Summary Statistics for Numerical Features:\n", summary_stats)

# Calculate frequency distributions for categorical features
categorical_features = ['airline', 'source_city', 'destination_city', 'class', 'stops']
for col in categorical_features:
    frequency_distribution = df[col].value_counts()
    print(f"\nFrequency Distribution for '{col}':\n{frequency_distribution}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histograms for numerical features
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
sns.histplot(df['duration'], kde=True)
plt.title('Distribution of Flight Duration')
plt.xlabel('Duration (hours)')

plt.subplot(1, 3, 2)
sns.histplot(df['days_left'], kde=True)
plt.title('Distribution of Days Left')
plt.xlabel('Days Left')

plt.subplot(1, 3, 3)
sns.histplot(df['price'], kde=True)
plt.title('Distribution of Price')
plt.xlabel('Price')

plt.tight_layout()
plt.show()

# Box plots for numerical features
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
sns.boxplot(y=df['duration'])
plt.title('Box Plot of Flight Duration')

plt.subplot(1, 3, 2)
sns.boxplot(y=df['days_left'])
plt.title('Box Plot of Days Left')

plt.subplot(1, 3, 3)
sns.boxplot(y=df['price'])
plt.title('Box Plot of Price')

plt.tight_layout()
plt.show()


# Comparisons across categorical features
categorical_features = ['airline', 'source_city', 'destination_city', 'class', 'stops']
numerical_features = ['duration', 'days_left', 'price']

for cat_feature in categorical_features:
    plt.figure(figsize=(15, 15))
    for i, num_feature in enumerate(numerical_features):
        plt.subplot(3, 1, i + 1)
        sns.boxplot(x=cat_feature, y=num_feature, data=df)
        plt.title(f'Box Plot of {num_feature} vs. {cat_feature}')
        plt.xticks(rotation=45, ha='right')
        plt.tight_layout()
    plt.show()

    plt.figure(figsize=(15,15))
    for i, num_feature in enumerate(numerical_features):
        plt.subplot(3,1,i+1)
        sns.violinplot(x=cat_feature, y=num_feature, data=df)
        plt.title(f"Violin Plot of {num_feature} vs. {cat_feature}")
        plt.xticks(rotation=45, ha='right')
        plt.tight_layout()
    plt.show()

    plt.figure(figsize=(15, 15))
    for i, num_feature in enumerate(numerical_features):
        plt.subplot(3, 1, i + 1)
        sns.barplot(x=cat_feature, y=num_feature, data=df, estimator=sum) # Use sum for barplot
        plt.title(f'Bar Plot of {num_feature} vs. {cat_feature}')
        plt.xticks(rotation=45, ha='right')
        plt.tight_layout()
    plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Select numerical features
numerical_features = ['duration', 'days_left', 'price']
numerical_df = df[numerical_features]

# Calculate the correlation matrix
correlation_matrix = numerical_df.corr()
print("Correlation Matrix:")
display(correlation_matrix)

# Create scatter plots
sns.pairplot(numerical_df)
plt.suptitle("Scatter Plots of Numerical Features", y=1.02)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Pair plot
sns.pairplot(df, vars=['duration', 'days_left', 'price'], hue='class')
plt.suptitle('Pairwise Relationships of Numerical Features Colored by Class', y=1.02)
plt.show()

# Heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(df[['duration', 'days_left', 'price']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix Heatmap')
plt.show()

# Grouped comparisons (box plots)
categorical_features = ['airline', 'stops']
numerical_features = ['price']
for cat_feature in categorical_features:
    plt.figure(figsize=(12, 6))
    sns.boxplot(x=cat_feature, y='price', hue='class', data=df)
    plt.title(f'Price Distribution by {cat_feature} and Class')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
```

```python
import pandas as pd

# Grouped mean price
grouped_price = df.groupby(['class', 'airline'])[['price', 'duration', 'days_left']].mean()
print("Mean price, duration, and days left grouped by class and airline:\n", grouped_price)

# Grouped mean price by stops and class
grouped_price_stops_class = df.groupby(['stops', 'class'])['price'].mean()
print("\nMean price grouped by stops and class:\n", grouped_price_stops_class)

#Further analysis by source and destination
grouped_price_src_dest = df.groupby(['source_city', 'destination_city'])[['price', 'duration', 'days_left']].mean()
print("\nMean price, duration, and days left grouped by source and destination cities:\n", grouped_price_src_dest)
```
